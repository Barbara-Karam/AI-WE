```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sympy import symbols, diff

# Ask the user for the function
f = input("Please enter a function of 'x' and 'y': ")

# Define the symbols
x, y = symbols('x y')

# Calculate the gradient
dfdx = diff(f, x)
dfdy = diff(f, y)

# Create a range of values for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid for x and y
X, Y = np.meshgrid(x, y)

# Vectorize the functions so they can operate on the numpy arrays X and Y
f_vectorized = np.vectorize(lambda x, y: eval(f))
dfdx_vectorized = np.vectorize(lambda x, y: eval(str(dfdx)))
dfdy_vectorized = np.vectorize(lambda x, y: eval(str(dfdy)))

# Calculate Z values for the function and its gradient
Z = f_vectorized(X, Y)
Zdx = dfdx_vectorized(X, Y)
Zdy = dfdy_vectorized(X, Y)

# Create a 3D plot
fig = plt.figure(figsize=(18, 6))

# Plot the function
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z)
ax1.set_title('Function')

# Plot the gradient
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X, Y, Zdx + Zdy)
ax2.set_title('Gradient')

# Plot the gradient contour
ax3 = fig.add_subplot(133)
ax3.quiver(X[::5], Y[::5], Zdx[::5], Zdy[::5])
ax3.set_title('Gradient Contour')

plt.show()
```
