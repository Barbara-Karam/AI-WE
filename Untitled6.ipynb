```python
def lagrange_points(m1, m2, d):
    # Calculate the position of the Lagrange points
    r1 = d * (1 - (m2 / (3*m1))**(1/3))
    r2 = d * (1 + (m2 / (3*m1))**(1/3))
    r3 = d * (-1 - (7*m2 / (12*m1))**(1/3))

    return r1, r2, r3

def main():
    # Masses and distance
    m1 = float(input("Enter the mass of the first body (in scientific notation, e.g., 5.974e24 for Earth): "))
    m2 = float(input("Enter the mass of the second body (in scientific notation, e.g., 7.348e22 for Moon): "))
    d = float(input("Enter the distance between the bodies (in meters): "))

    # Calculate Lagrange points
    L1, L2, L3 = lagrange_points(m1, m2, d)

    print(f"L1: {L1} m")
    print(f"L2: {L2} m")
    print(f"L3: {L3} m")

if __name__ == "__main__":
    main()
```
